```python
""" Here we are cleaning the data ,Basically 
Clean the data by removing the edges where the node metadata information
is not present. Simply prune any nodes that are not part of the associations
file and vice versa donâ€™t put any edges in the associations file if there is no
node metadata available."""
import pandas as pd

# Define the filter function
def filter_identifiers(items_list, key):
    """
    Extract identifiers from a list of dictionaries based on the specified key.

    Args:
        items_list (list): List of dictionaries containing identifiers.
        key (str): Key to extract identifiers.

    Returns:
        set: A set of identifiers for the specified key.
    """
    return {item[key] for item in items_list if key in item}

# Sample data: lists of dictionaries for diseases, proteins, and drugs

# Filter identifiers
disease_identifiers = filter_identifiers(diseases, "DISEASE")
protein_identifiers = filter_identifiers(proteins, "PROTEIN")
drug_identifiers = filter_identifiers(drugs, "DRUG")

# Combine all identifiers into one set
valid_identifiers = disease_identifiers | protein_identifiers | drug_identifiers

# Step 1: Read the associations.csv
association_file_path = "associations.csv"  # Replace with the actual path
df = pd.read_csv(association_file_path)

# Step 2: Filter associations based on valid_identifiers
filtered_df = df[
    (df['HEAD'].isin(valid_identifiers)) &
    (df['TAIL'].isin(valid_identifiers))
]

# Save the filtered associations
filtered_associations_path = "filtered_associations.csv"
filtered_df.to_csv(filtered_associations_path, index=False)
print("Filtered_associations saved as csv file")
print(f"Filtered associations saved to {filtered_associations_path}.")

# Step 3: Filter dictionaries based on relations in filtered associations
# Get unique HEAD and TAIL values in the filtered associations
unique_identifiers_in_associations = set(filtered_df['HEAD']).union(set(filtered_df['TAIL']))

# Filter diseases
filtered_diseases = [d for d in diseases if d['DISEASE'] in unique_identifiers_in_associations]

# Filter proteins
filtered_proteins = [p for p in proteins if p['PROTEIN'] in unique_identifiers_in_associations]

# Filter drugs
filtered_drugs = [d for d in drugs if d['DRUG'] in unique_identifiers_in_associations]

# Save the filtered dictionaries



def clean_data(proteins, drugs, diseases):
    # Function to clean invalid property values
    def sanitize_value(value):
        if isinstance(value, dict) and '$numberDouble' in value:
            return None  # Replace invalid maps with None
        return value

    # Clean proteins
    for protein in proteins:
        for key in protein:
            protein[key] = sanitize_value(protein[key])

    # Clean drugs
    for drug in drugs:
        for key in drug:
            drug[key] = sanitize_value(drug[key])

    # Clean diseases
    for disease in diseases:
        for key in disease:
            disease[key] = sanitize_value(disease[key])

    return proteins, drugs, diseases
filtered_proteins, filtered_drugs, filtered_diseases = clean_data(filtered_proteins, filtered_drugs, filtered_diseases)



with open("filtered_diseases.json", "w") as f:
    json.dump(filtered_diseases, f)

with open("filtered_proteins.json", "w") as f:
    json.dump(filtered_proteins, f)

with open("filtered_drugs.json", "w") as f:
    json.dump(filtered_drugs, f)
print("filtered_associations saved as csv file")
print("Filtered identifiers saved as JSON files.")

# Saving all the data as json files and filtered_ssociations as csv
```

```python
from neo4j import GraphDatabase
#Initialis the graph parametres and assiging some fucntions
class Neo4jManager:
    def __init__(self, uri, username, password):
        self.driver = GraphDatabase.driver(uri, auth=(username, password))

    def close(self):
        self.driver.close()

    def run_query(self, query, parameters=None):
        with self.driver.session() as session:
            return session.run(query, parameters)

    def get_graph_stats(self):
        query = """
        MATCH (n)
        WITH COUNT(n) AS total_nodes
        MATCH ()-[r]->()
        RETURN total_nodes, COUNT(r) AS total_edges;
        """
        with self.driver.session() as session:
            result = session.run(query)
            return result.single()

    def fetch_node_by_id(self, node_id):
        query = """
        MATCH (n)
        WHERE ID(n) = $node_id
        RETURN n
        """
        parameters = {"node_id": node_id}
        with self.driver.session() as session:
            result = session.run(query, parameters)
            return result.single()  # Return the node found


# Example usage:
neo4j_manager = Neo4jManager(url, username, password)

```

```python
#INSERTING THE DATA INTO THE GRAPH
def insert_nodes_and_relationships():
    # Insert Disease nodes
    for disease in filtered_diseases:
        query = """
        MERGE (d:Disease {id: $id})
        SET d += $properties
        """
        properties = {key: disease[key] for key in disease if key != 'DISEASE'}
        parameters = {"id": disease['DISEASE'], "properties": properties}
        neo4j_manager.run_query(query, parameters)

    # Insert Drug nodes
    for drug in filtered_drugs:
        query = """
        MERGE (d:Drug {id: $id})
        SET d += $properties
        """
        properties = {key: drug[key] for key in drug if key != 'DRUG'}
        parameters = {"id": drug['DRUG'], "properties": properties}
        neo4j_manager.run_query(query, parameters)

    # Insert Protein nodes
    for protein in filtered_proteins:
        query = """
        MERGE (p:Protein {id: $id})
        SET p += $properties
        """
        properties = {key: protein[key] for key in protein if key != 'PROTEIN'}
        parameters = {"id": protein['PROTEIN'], "properties": properties}
        neo4j_manager.run_query(query, parameters)

    # Insert Relationships based on the RELATION column
    for _, row in filtered_df.iterrows():
        relation = row['RELATION']
        head = row['HEAD']
        tail = row['TAIL']

        # Handle all specified relations dynamically
        if relation == 'DRUG_TARGET':
            query = """
            MATCH (d:Drug {id: $drug_id}), (p:Protein {id: $protein_id})
            MERGE (d)-[:DRUG_TARGET]->(p)
            """
            parameters = {"drug_id": head, "protein_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'DRUG_CARRIER':
            query = """
            MATCH (d:Drug {id: $drug_id}), (p:Protein {id: $protein_id})
            MERGE (d)-[:DRUG_CARRIER]->(p)
            """
            parameters = {"drug_id": head, "protein_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'DRUG_ENZYME':
            query = """
            MATCH (d:Drug {id: $drug_id}), (p:Protein {id: $protein_id})
            MERGE (d)-[:DRUG_ENZYME]->(p)
            """
            parameters = {"drug_id": head, "protein_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'DRUG_TRANSPORTER':
            query = """
            MATCH (d:Drug {id: $drug_id}), (p:Protein {id: $protein_id})
            MERGE (d)-[:DRUG_TRANSPORTER]->(p)
            """
            parameters = {"drug_id": head, "protein_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'DRUG_DISEASE_ASSOCIATION':
            query = """
            MATCH (d:Drug {id: $drug_id}), (dis:Disease {id: $disease_id})
            MERGE (d)-[:DRUG_DISEASE_ASSOCIATION]->(dis)
            """
            parameters = {"drug_id": head, "disease_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'PPI':
            query = """
            MATCH (p1:Protein {id: $protein1_id}), (p2:Protein {id: $protein2_id})
            MERGE (p1)-[:PPI]->(p2)
            """
            parameters = {"protein1_id": head, "protein2_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'DDI':
            query = """
            MATCH (d1:Drug {id: $drug1_id}), (d2:Drug {id: $drug2_id})
            MERGE (d1)-[:DDI]->(d2)
            """
            parameters = {"drug1_id": head, "drug2_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'PROTEIN_DISEASE_ASSOCIATION':
            query = """
            MATCH (p:Protein {id: $protein_id}), (d:Disease {id: $disease_id})
            MERGE (p)-[:PROTEIN_DISEASE_ASSOCIATION]->(d)
            """
            parameters = {"protein_id": head, "disease_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'DPI':
            query = """
            MATCH (p:Drug {id: $drug_id}), (d:Protein {id: $protein_id})
            MERGE (p)-[:DPI]->(d)
            """
            parameters = {"drug_id": head, "protein_id": tail}
            neo4j_manager.run_query(query, parameters)

        elif relation == 'SIMILAR_TO':
            query = """
            MATCH (d1:Drug {id: $drug1_id}), (d2:Drug {id: $drug2_id})
            MERGE (d1)-[:SIMILAR_TO]->(d2)
            """
            parameters = {"drug1_id": head, "drug2_id": tail}
            neo4j_manager.run_query(query, parameters)

# Insert data
insert_nodes_and_relationships()

print("Graph created successfully")
```

```python
#SAVING THE NODES AS JSON
import json
from neo4j import GraphDatabase

# Setup Neo4j connection


# Connect to Neo4j
driver = GraphDatabase.driver(url, auth=(username, password))

# Function to fetch all nodes
def fetch_nodes():
    with driver.session() as session:
        result = session.run("MATCH (n) RETURN n LIMIT 53212 ")  # Fetching first 100 nodes
        nodes = [{"id": record["n"].id, "data": dict(record["n"])} for record in result]
    return nodes

# Fetch nodes
nodes = fetch_nodes()

# Save nodes to JSON file
with open("nodes.json", "w") as file:
    json.dump(nodes, file, indent=4)

print(f"Nodes data saved to 'nodes.json'")
```

```python
#converting the nodes to embeddings 
from sentence_transformers import SentenceTransformer
import pickle

# Initialize the model
model = SentenceTransformer('all-MiniLM-L6-v2')  # 

def embed_nodes(nodes):
    # Safely extract descriptions and convert to strings
    node_descriptions = []
    for node in nodes:
        description = node.get('data', '')  # Get the 'data' field or default to empty string
        if isinstance(description, str):  # Check if it's already a string
            node_descriptions.append(description)
        elif isinstance(description, (int, float)):  # Convert int/float to string
            node_descriptions.append(str(description))
        else:
            node_descriptions.append('')  # Handle any other non-string values

    # Generate embeddings
    embeddings = model.encode(node_descriptions)

    # Save embeddings to a file
    with open('embeddings.pkl', 'wb') as f:
        pickle.dump(embeddings, f)

    return embeddings

embeddings = embed_nodes(nodes)
```

```python
#ADDING THE EMBEDDINGS to the vector database
import chromadb 

client = chromadb.Client()

# Create or load a collection
collection = client.create_collection("neo4j_nodes")  # Create a collection for the embeddings
def store_embeddings_in_batches(embeddings, nodes, batch_size=10):
    for i in range(0, len(nodes), batch_size):
        batch_nodes = nodes[i:i + batch_size]
        batch_embeddings = embeddings[i:i + batch_size]
        for node, embedding in zip(batch_nodes, batch_embeddings):
            document = json.dumps(node['data']) if isinstance(node['data'], dict) else str(node['data'])
            collection.add(
                ids=[node['data']['id']],
                documents=[document],
                embeddings=[embedding]
            )
store_embeddings_in_batches(embeddings, nodes)
print("Embeddings added successfully")
```

```python
#FUNCTION TO CREATE THE EMBEDDING FOR THE QUERY 
def query_collection_with_embedding(collection, query_text, model, n_results=2):
    """
    Function to query a collection using a generated embedding for the provided query text.
    
    Parameters:
        collection: The collection object to query.
        query_text (str): The query text to generate an embedding for.
        model: The model used for generating embeddings.
        n_results (int): The number of results to retrieve (default is 2).
    
    Returns:
        list: A list of query results.
    """
    def generate_query_embedding(query_text):
        """Generates embedding for a given query text."""
        return model.encode([query_text])[0]
    
    # Perform the query with the generated embedding
    results = collection.query(
        query_embeddings=generate_query_embedding(query_text),
        n_results=n_results
    )
    
    return results

# Example usage:

```

```python
# CONTIINUOS CHATBOT
import spacy
from rapidfuzz import fuzz, process  # For fuzzy matching

class ContinuousChatbot:
    def __init__(self, chain, nlp_model="en_core_web_md"):
        self.chain = chain
        self.previous_context = ""
        self.nlp = spacy.load(nlp_model)  # Load a robust NLP model

        # Standardized disease terms and their variations
        self.standard_diseases = {
            "Tuberculosis": ["tb", "tuberculoses", "tubercular", "Tuberculosis", "tuberculosis"],
            "Ovarian Cancer": ["ovarian cancer", "ovarian carcinoma", "ovarian neoplasm", "Ovarian cancer", "ovarian cancer"],
            "Breast Cancer": ["breast cancer", "mammary carcinoma", "breast neoplasm", "Breast cancer", "breast carcinoma"],
            "Diabetes Mellitus": ["diabetes", "diabetes mellitus", "type 1 diabetes", "type 2 diabetes", "diabetes type 2", "diabetes mellitus", "DM"],
            "Asthma": ["asthma", "bronchial asthma", "reactive airway disease", "asthma", "bronchial asthma"],
            "HIV/AIDS": ["HIV", "aids", "human immunodeficiency virus", "acquired immunodeficiency syndrome", "HIV/AIDS"],
            "Hypertension": ["hypertension", "high blood pressure", "HBP", "hypertensive disorder", "hypertension"],
            "Parkinson's Disease": ["parkinson's disease", "parkinsonism", "PD", "parkinson disease", "Parkinson's disease"],
            "Alzheimer's Disease": ["alzheimer's disease", "alzheimer disease", "AD", "alzheimers", "Alzheimer's disease"],
            "Influenza": ["influenza", "flu", "seasonal flu", "swine flu", "H1N1", "influenza virus"],
            "Malaria": ["malaria", "plasmodium infection", "malaria fever", "malaria"],
            "Cystic Fibrosis": ["cystic fibrosis", "CF", "cystic fibrosis disorder", "cystic fibrosis disease"],
            "Sickle Cell Anemia": ["sickle cell anemia", "sickle cell disease", "sickle cell disorder", "SCD", "sickle cell"],
            "Tuberculous Meningitis": ["tuberculous meningitis", "TB meningitis", "meningitis tuberculosa"],
            "Chronic Obstructive Pulmonary Disease": ["COPD", "chronic obstructive pulmonary disease", "emphysema", "chronic bronchitis"],
            "Multiple Sclerosis": ["multiple sclerosis", "MS", "sclerosis multiple", "multiple sclerosis disease"],
            "Epilepsy": ["epilepsy", "seizure disorder", "convulsive disorder", "epileptic seizure", "epilepsy"],
            "Leukemia": ["leukemia", "blood cancer", "leukemia", "leukemic", "leukemia cancer"],
            "Lung Cancer": ["lung cancer", "pulmonary carcinoma", "lung carcinoma", "lung neoplasm", "non-small cell lung cancer"],
            "Renal Failure": ["renal failure", "kidney failure", "chronic kidney disease", "CKD", "renal dysfunction"],
            "Peptic Ulcer Disease": ["peptic ulcer", "gastric ulcer", "stomach ulcer", "peptic ulcer disease", "gastric ulcer disease"],
            "Anemia": ["anemia", "iron deficiency anemia", "aplastic anemia", "sickle cell anemia", "hemolytic anemia"],
            "Psoriasis": ["psoriasis", "psoriatic skin disease", "skin psoriasis", "psoriatic arthritis"],
        }


        # Define fallback phrases for semantic similarity checking
        self.fallback_phrases = [
            "I don't know the answer",
            "I'm not sure",
            "I don't have enough information",
            "I'm sorry, I can't help with that"
        ]
        # Convert fallback phrases to spaCy Doc objects for similarity comparison
        self.fallback_docs = [self.nlp(phrase) for phrase in self.fallback_phrases]

    # def format_context1(self, context):
    #     if isinstance(context, list):
    #         formatted_context = [
    #             item['p.name'] for item in context if isinstance(item, dict) and 'p.name' in item and item['p.name'] is not None
    #         ]
    #         if formatted_context:
    #             return ", ".join(formatted_context)
    #         else:
    #             return ""
    #     return "No additional context available."

    def format_context(self, context, user_query):
        instructions = (
            "You are extracting the name of diseases, drugs, genes or proteins, exposures, molecular function, "
            "pathways, phenotypes, cellular components, biological processes entities from the text. "
            "You are extracting multi-word entities (like the name of pathways or gene names) from the text. "
            "Ensure to capture complete and meaningful phrases, and avoid partial entity identification. "
            "Do semantic searching to understand context and relationships."
        )
        formatted_context_query = f"{instructions} {user_query} {context}"
        response = self.chain.invoke({"query": formatted_context_query})
        return response.get("result", "No relevant context available.")


    def format_context2(self, context):
        print("context:", context)

        # Define the instruction to make the model extract important terms
        formatted_context_query = f"List out the important terms: {context}"

        for attempt in range(3):  # Retry up to 3 times if fallback is detected
            print(f"Attempt {attempt + 1} for format_context2...")

            try:
                # Pass the formatted context to the chain for processing
                response = self.chain.invoke({"query": formatted_context_query})

                # Extract and clean the result text
                result_text = response.get("result", "").strip()

                # Check if the result is semantically close to a fallback message
                if not result_text or self.is_fallback_message(result_text):
                    print(f"Response deemed fallback on attempt {attempt + 1}: '{result_text}'")
                    continue  # Retry if it's a fallback message

                # If a valid response is found, return it
                return result_text

            except Exception as e:
                print(f"Attempt {attempt + 1} failed with error: {e}")

        # If all attempts fail, return a default message
        return "No relevant context available after three attempts."




    def nlp_process(self, user_query):
        """Apply NLP to standardize disease terms in the query."""
        doc = self.nlp(user_query)

        # Replace terms with the closest standardized disease names
        processed_tokens = [
            self.find_closest_standard_term(token.text) for token in doc
        ]

        return " ".join(processed_tokens)

    def find_closest_standard_term(self, token_text):
        """Find the closest matching disease name using fuzzy matching."""
        token_text = token_text

        # Check all disease names and their variations
        for standard_name, variations in self.standard_diseases.items():
            variations = [standard_name] + variations

            # Use fuzzy matching to find the closest match
            result = process.extractOne(token_text, variations, scorer=fuzz.partial_ratio)

            if result:
                match, score, *_ = result  # Unpack additional values safely
                if score >= 80:  # Threshold for accepting a close match
                    return standard_name  # Return the standardized name

        return token_text  # Return the original token if no close match is found

    def is_fallback_message(self, result_text):

        """Check if the result text is a fallback message by comparing semantic similarity."""
        result_doc = self.nlp(result_text)

        # Compare result_doc with each fallback_doc, return True if any similarity is above the threshold
        for fallback_doc in self.fallback_docs:
            if result_doc.similarity(fallback_doc) >= 0.75:  # Threshold can be adjusted
                return True
        return False



    def ask(self, user_query):
        enhanced_query = self.nlp_process(user_query)
        #full_query = f"{self.previous_context}\n{enhanced_query}"
        full_query = f"{enhanced_query}"

        for attempt in range(3):  # Retry up to 3 times if context is empty
            print(f"Attempt {attempt + 1}...")
            try:
                # Pass the enhanced query to the chain for processing
                response = self.chain.invoke({"query": user_query})

                result_text = response["result"].strip()

                # If the result is a fallback message, process the intermediate context
                if self.is_fallback_message(result_text):
                    intermediate_steps = response.get("intermediate_steps", [])
                    context_data_list = [step["context"] for step in intermediate_steps if "context" in step]
                    print("context_data_list",context_data_list)

                    if context_data_list:
                        # Pass context to the LLM for natural language formatting
                        all_contexts = [self.format_context2(data) for data in context_data_list]
                        combined_readable_context = "\n".join(all_contexts)
                        print("combined_readable_context",combined_readable_context)
                        response["result"] = f"\n{combined_readable_context}"

                # If the result is non-empty, append it to the previous context
                if response["result"].strip():
                    self.previous_context += f"\n{response['result']}"
                    response['result']=self.format_context2(response['result'])
                    return response



            except Exception as e:
                print(f"Attempt {attempt + 1} failed with error: {e}")

        # If all attempts fail or no context is found, return a fallback response
        return {"result": "I don't know the answer", "intermediate_steps": []}

    def chat(self):
        """Start the chatbot conversation."""
        print("Welcome to the Medical Knowledge Chatbot! ")
        while True:
            user_input = input("You: ")
            print("You: ",user_input)
            if user_input.lower() == 'exit':
                print("Goodbye!")
                break

            response = self.ask(user_input)
            print("CHATBOT:" ,self.format_context2(response['result']))
            #print(f"\nChatbot: {response['result']}")

            # intermediate_steps = response.get("intermediate_steps", [])
            # context_data_list = [
            #     step["context"] for step in intermediate_steps if "context" in step
            # ]

            # if context_data_list:
            #     print("\n[Context Extracted]:")
            #     for context_data in context_data_list:
            #         print(self.format_context(context_data))
            # else:
            #     print("[No Relevant Context Available]")
```
